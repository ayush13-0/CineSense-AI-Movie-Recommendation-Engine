# üîÑ Converting JSON-like Columns
üí° Statement:

Convert the genres, keywords, production companies, and cast into lists of strings.

```python
import ast

def convert(obj):
    L = []
    obj = ast.literal_eval(obj)  # Convert string to list of dictionaries
    for i in obj:
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

üëÅÔ∏è Review of Data after Execution

```python
movies.head()
```

# üë• Extracting Top 3 Cast Members
üí° Statement:

Extract only the top 3 cast members from the cast column.

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

movies['cast'] = movies['cast'].apply(convert3)
```

üëÅÔ∏è Review of Data after Execution

```python
movies.head()
```

# üé¨ Processing the Crew Column
üí° Statement:
Extract the director‚Äôs name from the crew field.

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(fetch_director)
```

üëÅÔ∏è Review of Data after Execution

```python
movies.head()
```

# ‚úçÔ∏è Tokenizing Overview
üí° Statement: Split movie descriptions into individual tokens.

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

üëÅÔ∏è Review of Data after Execution

# üî§ 14. Removing Spaces from Multi-word Names
üí° Statement: Fix multi-word strings like ‚ÄúScience Fiction‚Äù ‚Üí ‚ÄúScienceFiction‚Äù.

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

üëÅÔ∏è Review of Data after Execution

```python
movies.head()
```

# üß¨ 19. Text Vectorization Using Bag-of-Words
üí° Statement: Use CountVectorizer to convert text into numerical vectors.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

# üßπ Text Preprocessing with Stemming
üìå Importing Porter Stemmer & Creating Function
- üìé The stemmer reduces words to their base form (e.g., "loved" ‚Üí "love")
- üìé Helps improve similarity between movies with related descriptions.

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

# üìå Demonstration of Stemming
üìé Shows how stemmed text becomes shorter and more uniform.

# üìå Applying Stemming to the Entire Dataset
üìé Cleans all movie tags for accurate vector creation.

```python
df['tags'] = df['tags'].apply(stem)
```

# üìè Compute Cosine Similarity
üí° Statement: Calculate similarity scores between all movie vectors.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

# üéØ Recommendation Function
üí° Statement: Define a function that lists 5 closest movies based on similarity.

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

# üß™Testing the Recommendation System
üí° Statement: Test recommendations for a sample movie title.

```python
recommend('Avatar')
```

```python
recommend("Ender's Game")
```

```python
recommend("Lost in Space")
```
